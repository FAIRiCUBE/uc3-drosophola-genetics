```python
import os
import pandas as pd
from smart_open import smart_open

aws_key = os.environ['S3_FAIRICUBE_STORAGE_KEY']
aws_secret = os.environ['S3_FAIRICUBE_STORAGE_SECRET']

bucket_name = 'fairicube'
object_key = 'genomics_data/raw/filtered/Europe_genomewide_freq.csv'

path = 's3://{}:{}@{}/{}'.format(aws_key, aws_secret, bucket_name, object_key)

df = pd.read_csv(smart_open(path), sep="\t")
```
